```python
def caseFolding(string):
    return string.lower()

def cleaningData(string):
#     remove username
    string = ' '.join(re.sub("([@#][A-Za-z0-9_]+)|(\w+:\/\/\S+)"," ", string).split())
#     remove hyperlink
    string = re.sub(r'^https?:\/\/.*[\r\n]*', ' ', string, flags=re.MULTILINE)
#     remove hashtag
#     remove numbers
    string = s = re.sub("\d+", "", string)
#     remove special character
    string = re.sub('[^A-Za-z0-9]+', ' ', string)
#     remove first and back whitespace
    return string.strip()

def getunigram(string):
    tokens = [token for token in string.split(" ") if token != ""]
    return tokens

def getbigram(string):
    tokens = [token for token in string.split(" ") if token != ""]
    ngrams = zip(*[tokens[i:] for i in range(2)])
    return [" ".join(ngram) for ngram in ngrams]

def gettrigram(string):
    tokens = [token for token in string.split(" ") if token != ""]
    ngrams = zip(*[tokens[i:] for i in range(3)])
    return [" ".join(ngram) for ngram in ngrams]

def generate_dictionary(list_token, alltoken):
    kamuskata={}
    for index in alltoken:
        kamuskata[index] = 0

    for index in list_token:
        if index in kamuskata.keys():
            kamuskata[index]+=1
        else:
            kamuskata[index]+= 0
    return kamuskata

def join_list(listinlist):
    return (list(chain(*listinlist)))

def countWordinClass(kata, list_dokumen_kelas):
    kata = 0
    for index in list_dokumen_kelas:
        for word in index:
            if kata in word:
                kata +=1
    return kata

def countWords(dic):
    return sum(dic.values())

def countTerm(kamuskata):
    return len(kamuskata)

def wordProbability(value_kata, total_nilai_kata, total_term):
    return ( value_kata + 1 )/(total_nilai_kata+total_term)
```

```python
def getlikelihood(kamus_kelas,kamus):
    likelihood = {}
    for key in kamus_kelas:
        likelihood[key] = wordProbability(kamus_kelas[key],countWords(kamus_kelas), countTerm(kamus))
    return likelihood
```

```python
def getPosterior(query, likelihood):
    posterior = 1
    for index in query:
        if index in likelihood.keys():
            posterior *= likelihood.get(index)
    return posterior/2
```

```python
def getPrediction(pos, neg):
    label = ''
    if pos > neg:
        label = 'positif'
    else:
        label = 'negatif'
    return label
```

```python
def getpos(query, pos, neg):
    post = []
    nega = []
    for i in range(len(query)):
        post.append(getPosterior(query[i], pos))
        nega.append(getPosterior(query[i], neg))
    return post, nega
```

```python
def confussion_matrix(list_labeluji , list_hasilprediksi):
    TP = 0
    FP = 0
    TN =0
    FN =0
    for index in range(len(list_labeluji)):
        if list_labeluji[index] == list_hasilprediksi[index]:
            if list_labeluji[index]=="positif":
                TP+=1
            else:
                TN+=1
        else:
            if list_labeluji[index] =="positif":
                FN+=1
            else:
                FP+=1
    return TP, TN, FP, FN
```

```python
def evaluasi(TP,TN,FP,FN):
    try:
        accuracy = (TP+TN)/(TP+FP+FN+TN)
    except ZeroDivisionError:
        accuracy = 0
    try:
        precision = (TP)/(TP+FP)
    except ZeroDivisionError:
        precision = 0
    try:
        recall = (TP)/(TP+FN)
    except ZeroDivisionError:
        recall = 0
    try:
        Fmeasure = 2 *precision*recall/(precision+recall)
    except ZeroDivisionError:
        Fmeasure = 0
    return accuracy, precision, recall, Fmeasure
```
